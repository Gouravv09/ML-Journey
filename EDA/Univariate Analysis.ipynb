```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/train.csv')
```

```python
df.head()
```

# 1. Categorical Data

### a. Coutplot

```python
sns.countplot(df['Embarked']) # Create a count plot for the 'Embarked' column
```

### b. Piechart

```python
df['Sex'].value_counts().plot(kind='pie', autopct = '%.2f') # Create a pie chart for the 'Sex' column
```

# 2. Numerical Data

### a. Histogram

```python
sns.histplot(df['Age'], bins = 5) # Create a histogram for 'Age' using seaborn
```

```python
plt.hist(df['Age'], bins=5) # Create a histogram for 'Age' using matplotlib
```

### b. Distplot

```python
sns.histplot(df['Age']) # Visualize the distribution of 'Age' (updated from distplot)
```

### c. Boxplot

```python
sns.boxplot(df['Age']) # Create a box plot for 'Age' to show distribution and outliers
```

```python
df['Age'].mean() # Calculate the mean age
```
